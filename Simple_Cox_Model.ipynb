```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rc('xtick', labelsize=14)
mpl.rc('ytick', labelsize=14)
mpl.rc('axes', linewidth=1.5)
```

```
data_link = 'https://raw.githubusercontent.com/AilingLiu/Survival_analysis/master/Data/simulated%20HF%20mort%20data%20for%20GMPH%20(1K)%20final.csv'

hf = pd.read_csv(data_link, index_col='id')
hf.head()
```

Using Ethnicgroup as predictor to build simple cox model

```
hf.info()
```

```
df = hf.dropna(subset=['ethnicgroup'])
sum(df.ethnicgroup.isnull())
```

```
df.ethnicgroup.describe()
```

```
#!pip install lifelines
```

```
# convert the ethnicgroup into one hot encoding
ethnic = pd.get_dummies(df['ethnicgroup'], prefix='ethnic', drop_first=True)
ethnic.head()
```

```
from lifelines import CoxPHFitter
cox = CoxPHFitter()

data1=pd.concat([ethnic, df[['death', 'fu_time']]], axis=1)
model = cox.fit(data1, 'fu_time', event_col='death')
model.print_summary()
```
